```python
import pandas as pd
# 讀csv儅
df = pd.read_csv("Total Bond Market ETF List (82).csv",index_col='Symbol')
```

```python
# 去除2015-12-31之後的etf
filtered_df=df[df['Inception']<='2015-12-31']
```

```python
# 得到要抓數據的symbols
etf_symbols = filtered_df.index[:]
etf_symbols
```

```python
import requests
import re
import pandas as pd
from io import StringIO
import time

symbol = 'AGG'
#開始與結束時間
start_date = '1451491200'
end_date = '1552924800'
#使用regular expression找到對應的crumble
crumble_link = 'https://finance.yahoo.com/quote/{0}/history?p={0}'
crumble_regex = r'CrumbStore":{"crumb":"(.*?)"}'
cookie_regex = r'set-cookie: (.*?);'

link = crumble_link.format(symbol)
session = requests.Session()
response = session.get(link)

text = str(response.content)
match = re.search(crumble_regex,text)
crumbs = match.group(1)

cookie = session.cookies.get_dict()

url = 'https://query1.finance.yahoo.com/v7/finance/download/{0}?period1={1}&period2={2} \
&interval=1d&events=history&crumb={3}'.format(symbol,start_date,end_date,crumbs)
# 使用requests.get()得到數據
download=requests.get(url,cookies=cookie,timeout=5)
time.sleep(2)
# 讀csv檔案
df = pd.read_csv(StringIO(download.text)).set_index('Date')
# 調整col的名字
df = df[['Adj Close']]
df = df.rename(columns={'Adj Close':symbol})
```

```python
df.head(20)
```

```python
# 將symbols中的symbol對於的price都抓下來
for symbol in etf_symbols[1:]:
    link = crumble_link.format(symbol)
    session = requests.Session()
    response = session.get(link)

    text = str(response.content)
    match = re.search(crumble_regex,text)
    crumbs = match.group(1)

    cookie = session.cookies.get_dict()

    url = 'https://query1.finance.yahoo.com/v7/finance/download/{0}?period1={1}&period2={2} \
    &interval=1d&events=history&crumb={3}'.format(symbol,start_date,end_date,crumbs)
    
    try:
        download=requests.get(url,cookies=cookie,timeout=20)
        download.raise_for_status()
        time.sleep(5)

        new_df = pd.read_csv(StringIO(download.text)).set_index('Date')
        new_df = new_df[['Adj Close']]
        new_df = new_df.rename(columns={'Adj Close':symbol})
        df[symbol] = new_df;
    except:
        print(symbol+' failed.')
        pass
```

```python
# 對於失敗的symbol再抓一次
failed_symbols = ['BNDX','IAGG','ULST','FLTB','MINC','AGND']
for symbol in failed_symbols:
    link = crumble_link.format(symbol)
    session = requests.Session()
    response = session.get(link)

    text = str(response.content)
    match = re.search(crumble_regex,text)
    crumbs = match.group(1)

    cookie = session.cookies.get_dict()

    url = 'https://query1.finance.yahoo.com/v7/finance/download/{0}?period1={1}&period2={2} \
    &interval=1d&events=history&crumb={3}'.format(symbol,start_date,end_date,crumbs)
    
    try:
        download=requests.get(url,cookies=cookie,timeout=20)
        download.raise_for_status()
        time.sleep(5)

        new_df = pd.read_csv(StringIO(download.text)).set_index('Date')
        new_df = new_df[['Adj Close']]
        new_df = new_df.rename(columns={'Adj Close':symbol})
        df[symbol] = new_df;
    except:
        print(symbol+' failed.')
        pass
```
